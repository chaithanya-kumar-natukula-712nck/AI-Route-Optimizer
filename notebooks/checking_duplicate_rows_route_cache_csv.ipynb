```python
import sys
import os
import pandas as pd
```

```python
# Project root
project_root = os.path.abspath("..")

# Add to Python path
sys.path.append(project_root)
```

```python
# Route cache path
output_dir = os.path.join(project_root, "data")
input_path = os.path.join(output_dir, "route_cache.csv")
output_path = os.path.join(output_dir, "cleaned_route_cache.csv")
```

```python
# Load CSV
trips_df = pd.read_csv(input_path)

print(trips_df.head())


```

```python
trips_df.duplicated().sum()
```

```python
# --- Your duplicate removal logic goes here ---
cleaned_trips_df = trips_df.drop_duplicates() 
```

```python
os.makedirs(output_dir, exist_ok=True)

# Save file
cleaned_trips_df.to_csv(output_path, index=False)
```
