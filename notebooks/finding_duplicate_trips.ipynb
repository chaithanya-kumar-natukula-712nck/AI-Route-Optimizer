```python
def find_duplicates(matrix):
    seen = set()
    duplicates = set()
    
    for lst in matrix:
        # Convert to tuple because lists cannot be hashed in a set
        list_as_tuple = tuple(lst)
        if list_as_tuple in seen:
            duplicates.add(list_as_tuple)
        else:
            seen.add(list_as_tuple)
            
    # Convert back to a list of lists
    return [list(t) for t in duplicates]

duplicates=(find_duplicates(all_trips))
print(duplicates)
```

```python
from collections import defaultdict

def find_duplicate_indices(matrix):
    index_map = defaultdict(list)
    
    for index, lst in enumerate(matrix):
        # Convert to string to safely bypass any hidden nested structures
        list_as_string = "|".join(map(str, lst))
        index_map[list_as_string].append(index)
        
    # FIX: Convert the split string into a tuple so it can be a valid dict key
    duplicates = {tuple(k.split('|')): v for k, v in index_map.items() if len(v) > 1}
    return duplicates

# Run this on your data
duplicates_with_indices = find_duplicate_indices(all_trips)
print(duplicates_with_indices)
```

```python
all_duplicates_indices=[]
```

```python
for i in duplicates_with_indices.values():
    for j in i:
        all_duplicates_indices.append(j)
    all_duplicates_indices.pop(-1)
```
